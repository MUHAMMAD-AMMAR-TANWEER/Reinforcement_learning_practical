# 1. Importing Dependencies

```python
import os 
import gym #for building environments
from stable_baselines3 import A2C#Selected a one basic model for it 
from stable_baselines3.common.vec_env import VecFrameStack#mostly open Ai gym allows you to train on multiple environment right now with 4 envs
from stable_baselines3.common.evaluation import evaluate_policy #for Evaluation of Policy
from stable_baselines3.common.env_util import  make_atari_env
```

# 2. Test Environment

```python
#we have lot of dependencies issue but don't worry we have all things in requirements.txt
```

```python
env = gym.make('Breakout-v0')
```

```python
episodes = 5
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0 
    
    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)

        score+=reward
    print('Episode:{} Score:{}'.format(episode, score))
env.close()
```

# Vectorizing Environment and Training Model

```python
#Vectorization means simply run multiple models at same time to learn quickly via interactions
#it allows to train agent faster by training in parallet
```

```python
env = make_atari_env('Breakout-v0', n_envs=4, seed=0)
env = VecFrameStack(env, n_stack=4)
```

```python
env.reset()
```

```python
#env.render()  to show we have four simulated environment
```

```python
env.close()
```

```python


log_path = os.path.join('Training', 'Logs')

model = A2C('CnnPolicy', env, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=400000)
```

# Saving Model

```python
a2c_path = os.path.join('Training', 'Saved Models','A2C_Breakout_Model')
model.save(a2c_path)
```

```python
del model
```

```python
model = A2C.load(a2c_path, env)
```


# 5. Evaluate and Test

```python
env = make_atari_env('Breakout-v0', n_envs=1, seed=0)#we make this because evaluation of environment requires 
env = VecFrameStack(env, n_stack=4)
```

```python
evaluate_policy(model, env, n_eval_episodes=10 ,render=True)
```
